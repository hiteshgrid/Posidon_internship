```python
def assign_risk(row):
    # High income + low debt = Aggressive
    if "1Cr" in str(row['Annual Income Bucket']) and "0-5L" in str(row['Debt Bucket']):
        return "Aggressive"
    # Low income + high debt = Conservative
    elif "0-5L" in str(row['Annual Income Bucket']) and "20L" in str(row['Debt Bucket']):
        return "Conservative"
    else:
        return "Moderate"

df['Risk Appetite'] = df.apply(assign_risk, axis=1)
```

```python
# ============================================================
# FEATURE ENGINEERING
# ============================================================

from sklearn.base import BaseEstimator, TransformerMixin

# ============================================================
# CUSTOM FEATURE ENGINEERING CLASS
# ============================================================

class FinancialFeatureEngineering(
    BaseEstimator,
    TransformerMixin
):

    def __init__(self):

        # ====================================================
        # INCOME MAPPING
        # ====================================================

        self.income_mapping = {

            '₹0-5L': 250000,

            '₹5-10L': 750000,

            '₹10-15L': 1250000,

            '₹15-20L': 1750000,

            '₹20-25L': 2250000,

            '₹25-30L': 2750000,

            '₹35-40L': 3750000,

            '₹50-55L': 5250000,

            '₹1-1.5Cr': 12500000,

            '₹3Cr+': 30000000
        }

        # ====================================================
        # DEBT MAPPING
        # ====================================================

        self.debt_mapping = {

            'Not Applicable': 0,

            '₹2-4L': 300000,

            '₹4-6L': 500000,

            '₹6-8L': 700000,

            '₹8-10L': 900000,

            '₹10-12L': 1100000,

            '₹12-14L': 1300000,

            '₹14-16L': 1500000,

            '₹18-20L': 1900000,

            '₹20-25L': 2250000,

            '₹30-35L': 3250000
        }

        # ====================================================
        # NET WORTH MAPPING
        # ====================================================

        self.networth_mapping = {

            '₹0-5L': 250000,

            '₹10-15L': 1250000,

            '₹15-20L': 1750000,

            '₹25-30L': 2750000,

            '₹50-55L': 5250000,

            '₹1-1.5Cr': 12500000,

            '₹2-2.5Cr': 22500000,

            '₹3-3.5Cr': 32500000,

            '₹5-5.5Cr': 52500000,

            '₹8-8.5Cr': 82500000,

            '₹9-9.5Cr': 92500000,

            '₹20-20.5Cr': 202500000,

            '₹25Cr+': 300000000
        }

    # ========================================================
    # FIT
    # ========================================================

    def fit(self, X, y=None):

        return self

    # ========================================================
    # TRANSFORM
    # ========================================================

    def transform(self, X):

        X = X.copy()

        # ====================================================
        # CONVERT BUCKETS TO NUMBERS
        # ====================================================

        X['income_numeric'] = (

            X['Annual Income Bucket']

            .map(self.income_mapping)

            .fillna(0)
        )

        X['debt_numeric'] = (

            X['Debt Bucket']

            .map(self.debt_mapping)

            .fillna(0)
        )

        X['networth_numeric'] = (

            X['Net Worth Bucket']

            .map(self.networth_mapping)

            .fillna(0)
        )

        # ====================================================
        # FEATURE ENGINEERING
        # ====================================================

        # Debt To Income Ratio

        X['debt_to_income_ratio'] = (

            X['debt_numeric'] /

            (X['income_numeric'] + 1)
        )

        # Net Worth Minus Debt

        X['networth_minus_debt'] = (

            X['networth_numeric'] -

            X['debt_numeric']
        )

        # Financial Stability Score

        X['financial_stability_score'] = (

            X['networth_numeric'] +

            X['income_numeric'] -

            X['debt_numeric']
        )

        # ====================================================
        # REMOVE ORIGINAL BUCKET COLUMNS
        # ====================================================

        X.drop(

            columns=[

                'Annual Income Bucket',

                'Debt Bucket',

                'Net Worth Bucket'
            ],

            inplace=True
        )

        return X

# ============================================================
# APPLY FEATURE ENGINEERING
# ============================================================

feature_engineering = FinancialFeatureEngineering()

X = feature_engineering.fit_transform(X)

# ============================================================
# DISPLAY NEW FEATURES
# ============================================================

print("===================================")
print("FEATURE ENGINEERING COMPLETED")
print("===================================")

print("\nNew Dataset Shape:")
print(X.shape)

print("\nColumns After Feature Engineering:")
print(X.columns.tolist())

print("\nFirst 5 Rows:")
print(X.head())
```

```python
# ============================================================
# STEP 12 — FINAL PREPROCESSING PIPELINE
# ============================================================

from sklearn.pipeline import Pipeline

from sklearn.compose import ColumnTransformer

from sklearn.impute import SimpleImputer

from sklearn.preprocessing import (
    OneHotEncoder,
    StandardScaler
)

# ============================================================
# NUMERIC TRANSFORMER
# ============================================================

numeric_transformer = Pipeline(steps=[

    # Handle Missing Values

    (
        "imputer",

        SimpleImputer(
            strategy="median"
        )
    ),

    # Scale Numeric Features

    (
        "scaler",

        StandardScaler()
    )

])

# ============================================================
# CATEGORICAL TRANSFORMER
# ============================================================

categorical_transformer = Pipeline(steps=[

    # Handle Missing Values

    (
        "imputer",

        SimpleImputer(
            strategy="constant",
            fill_value="Unknown"
        )
    ),

    # One Hot Encoding

    (
        "encoder",

        OneHotEncoder(
            handle_unknown="ignore",
            sparse_output=False
        )
    )

])

# ============================================================
# FINAL COLUMN TRANSFORMER
# ============================================================

preprocessor = ColumnTransformer(

    transformers=[

        (
            "num",

            numeric_transformer,

            numeric_columns
        ),

        (
            "cat",

            categorical_transformer,

            categorical_columns
        )

    ]

)

# ============================================================
# DISPLAY
# ============================================================

print("===================================")
print("FINAL PREPROCESSING PIPELINE READY")
print("===================================")

print("\nNumeric Columns:")
print(numeric_columns)

print("\nCategorical Columns:")
print(categorical_columns)
```

```python
# ============================================================
# RE-SPLIT DATA AFTER FEATURE ENGINEERING
# ============================================================

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(

    X,

    y_encoded,

    test_size=0.2,

    random_state=42,

    stratify=y_encoded
)

print("===================================")
print("DATA RE-SPLIT COMPLETED")
print("===================================")

print("\nX_train Shape:")
print(X_train.shape)

print("\nX_test Shape:")
print(X_test.shape)
```

```python
# ============================================================
# STEP 13 — TRAIN ALL MODELS AGAIN
# ============================================================

from sklearn.metrics import (
    accuracy_score,
    classification_report
)

# ============================================================
# VARIABLES
# ============================================================

best_accuracy = 0

best_model = None

best_model_name = ""

results = []

print("===================================")
print("TRAINING ALL MODELS")
print("===================================")

# ============================================================
# LOOP THROUGH MODELS
# ============================================================

for name, algorithm in models.items():

    print(f"\nTraining: {name}")

    # ========================================================
    # CREATE PIPELINE
    # ========================================================

    model_pipeline = Pipeline(steps=[

        (
            "preprocessor",
            preprocessor
        ),

        (
            "classifier",
            algorithm
        )

    ])

    # ========================================================
    # TRAIN MODEL
    # ========================================================

    model_pipeline.fit(

        X_train,

        y_train
    )

    # ========================================================
    # PREDICT
    # ========================================================

    y_pred = model_pipeline.predict(

        X_test
    )

    # ========================================================
    # ACCURACY
    # ========================================================

    accuracy = accuracy_score(

        y_test,

        y_pred
    )

    print(f"Accuracy: {accuracy * 100:.2f}%")

    # ========================================================
    # STORE RESULTS
    # ========================================================

    results.append({

        "Model": name,

        "Accuracy": accuracy
    })

    # ========================================================
    # SAVE BEST MODEL
    # ========================================================

    if accuracy > best_accuracy:

        best_accuracy = accuracy

        best_model = model_pipeline

        best_model_name = name

# ============================================================
# RESULTS DATAFRAME
# ============================================================

results_df = pd.DataFrame(results)

results_df = results_df.sort_values(

    by="Accuracy",

    ascending=False
)

# ============================================================
# DISPLAY RESULTS
# ============================================================

print("\n===================================")
print("MODEL COMPARISON")
print("===================================")

print(results_df)

# ============================================================
# BEST MODEL
# ============================================================

print("\n===================================")
print("BEST MODEL")
print("===================================")

print(f"\nBest Model: {best_model_name}")

print(f"Best Accuracy: {best_accuracy * 100:.2f}%")

# ============================================================
# CLASSIFICATION REPORT
# ============================================================

best_predictions = best_model.predict(

    X_test
)

print("\n===================================")
print("CLASSIFICATION REPORT")
print("===================================")

print(

    classification_report(

        y_test,

        best_predictions
    )
)
```
